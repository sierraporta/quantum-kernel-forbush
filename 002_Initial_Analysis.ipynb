```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, Normalize
import seaborn as sns
import glob
import scienceplots
plt.style.use(['science','nature'])
plt.rc('axes', labelsize=11)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('legend', fontsize=11)
```

```python
datas,file=[],[]
for i in np.sort(glob.glob("Results/*.csv")):
    #print(i)
    dat=pd.read_csv(i)
    datas.append(dat)
    file.append(i)
```

```python
print("Realizados {} experimentos ({}%)".format(len(datas),np.round(len(datas)/(180+90)*100,2)))
df=pd.concat(datas)
df["File"]=file
print(f"Total filas: {len(df)}")  # debería ser 180

# Verificar duplicados
dups = df.duplicated(subset=["min_magn","n_qubits","reps","n_train_quantum"])
print(f"Duplicados: {dups.sum()}")  # debería ser 0

# Ver parámetros únicos
print(df["min_magn"].value_counts().sort_index())
print(df["n_train_quantum"].value_counts().sort_index())
df.to_csv("AllExperiments.csv")
```
